```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"E:\COLLEGE 2024\Python Repository\Python Pandas\Ice Cream Ratings.csv")
df = df.set_index('Date')
df
```

```python
# default: line graph
#  display data points connected by straight lines, showing trends and changes over time or continuous data.
df.plot(
    kind = 'line',
    #subplots = True,
    title = 'Ice Cream Ratings',
    xlabel = 'Daily Ratings',
    ylabel = 'Scores'
       )
plt.legend(loc='upper right', bbox_to_anchor=(1, 1),frameon = False )  # Top-right corner, slightly outside the plot
plt.show()
```

```python
# ploting a bar chart
df.plot(kind = 'bar')
```

```python
# ploting a stacked bar chart
# compare not only the total value across categories but also the contribution of individual components within those categories.
df.plot(kind = 'bar', stacked = True)
```

```python
# Bar chart: visualize and compare discrete categories of data
# ploting a single specific column
df['Flavor Rating'].plot(kind = 'bar')
```

```python
# Horizantal bar chart
df['Flavor Rating'].plot.barh()
```

```python
# scattor plot
# visualize the relationship or correlation between two numerical variables.
df.plot.scatter(x = 'Texture Rating', y = 'Overall Rating', s = 100, c = 'Green')
```

```python
# Histograms: distribution of a single variable by dividing the data into bins (intervals) and displaying the frequency of data points within each bin
df.plot.hist(bins = 20)
```

```python
# box plot
# statistical graph used to summarize the distribution of a dataset
# visual representation of data's central tendency, variability, and outliers.
df.boxplot()
```

```python
# Area
# combines elements of line charts and bar charts. 
# It is particularly useful for representing quantitative data and showing trends over time, 
# with an added emphasis on magnitude through the use of filled areas beneath the lines.
df.plot.area(figsize = (10,5))
```

```python
df.plot.pie(y = 'Flavor Rating',figsize=(9, 9))
```
