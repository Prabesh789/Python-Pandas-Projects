# Merge, Join, and Concatenate in Pandas

```python
import pandas as pd
```

```python
df1 = pd.read_csv(r"E:\COLLEGE 2024\Python Repository\Python Pandas\LOTR.csv")
```

```python
df1
```

```python
df2 = pd.read_csv(r"E:\COLLEGE 2024\Python Repository\Python Pandas\LOTR 2.csv")
```

```python
df2
```

```python
# Merge (inner join)
df1.merge(df2)
```

```python
# inner join
df1.merge(df2, how = 'inner')
```

```python
# FellowshipID: The common column used to merge.
# FirstName_x: The FirstName column from df1 is renamed to avoid conflicts.
# FirstName_y: The FirstName column from df2 is renamed to distinguish it from FirstName_x.
# Skills: From df1.
# Age: From df2.

# The join is performed based on the FellowshipID column.

df1.merge(df2, how = 'inner', on = 'FellowshipID')
```

```python
# FellowshipID: The column used in the join.
# FirstName: This column is common in both DataFrames and doesn’t need suffixes since it matches perfectly in the keys.

df1.merge(df2, how = 'inner', on = ['FellowshipID', 'FirstName'])
```

```python
df1.merge(df2, how = 'outer')
```

```python
df1.merge(df2, how = 'left')
```

```python
# Right join
```

```python
df1.merge(df2, how = 'right')
```

```python
df1.merge(df2, how = 'cross')
```

```python
# join functions requires more work 
df1.join(df2, on = 'FellowshipID', how = 'outer', lsuffix = '_Left', rsuffix = '_Right')
```

```python
# To work with join function, we can specify the index for df
df4 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'), lsuffix = '_Left', rsuffix = '_Right', how = 'outer')
df4
```

```python
pd.concat([df1,df2])
```

```python
# it concat the columns that are same
pd.concat([df1,df2], join = 'inner')
```

```python
# it concat all colums
pd.concat([df1,df2], join = 'outer')
```

```python
# it concat the columns that are same
pd.concat([df1,df2], join = 'outer', axis = 1)
```
